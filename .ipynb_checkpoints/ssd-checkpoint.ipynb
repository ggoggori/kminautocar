```python
def display_image(image):
  fig = plt.figure(figsize=(20, 15))
  plt.grid(False)
  plt.imshow(image)
# SSD를 거친 image 파일을 출력하기 위한 함수이다. Matplotlib 라이브러리를 사용했다.

def resize_image(filename, image_data, new_width=256, new_height=256,
                              display=False):
   '''
    Resize_image는 Image 파일 포맷을 SSD의 input으로 넣을 수 있게 전처리 하는 함수이다.
    PIL의 Image 메서드를 사용해서 image를 변수에 지정하고 RGB 형식으로 바꿔준다. 
    Return은 변환이 완료된 image파일의 주소이다.
    위에서는 ‘sample.JPG’ 를 입력으로 받아서 알맞은 변환을 가한 후 ‘convert.JPG’ 로 저장했다.

   ''' 
  pil_image = Image.open(image_data)
  pil_image = ImageOps.fit(pil_image, (new_width, new_height), Image.ANTIALIAS)
  pil_image_rgb = pil_image.convert("RGB")
  pil_image_rgb.save(filename, format="JPEG", quality=90)
  print("Image downloaded to %s." % filename)
  if display:#display 인자가 True이면 함수 실행시 결과물 출력해준다.
    display_image(pil_image)
  return filename


def draw_bounding_box_on_image(image,
                               ymin,
                               xmin,
                               ymax,
                               xmax,
                               color,
                               font,
                               thickness=4,
                               display_str_list=()):
    
    '''
    Draw_bounding_box_on_image 함수는 image 파일에 bounding box를 그려서 그 image를 반환하는 함수이다. 
    입력으로 받은 image에 PIL의 ImageDraw.Draw 메서드로 다른 입력인자인 ymin, xmin, ymax, xmax를 사용해서 line를 그린다. 
    이후에는 for문을 사용해 각 bounding box에 그려질 class명과 그 확률을 font로 나타낸다.

    '''
  draw = ImageDraw.Draw(image)
  im_width, im_height = image.size
  (left, right, top, bottom) = (xmin * im_width, xmax * im_width,
                                ymin * im_height, ymax * im_height) # 좌표값 구하기
  draw.line([(left, top), (left, bottom), (right, bottom), (right, top), # 구한 좌표값으로 line 그리기
             (left, top)],
            width=thickness,
            fill=color)

  display_str_heights = [font.getsize(ds)[1] for ds in display_str_list]
  total_display_str_height = (1 + 2 * 0.05) * sum(display_str_heights)

  if top > total_display_str_height: # 삽입할 text의 위치 설정
    text_bottom = top
  else:
    text_bottom = bottom + total_display_str_height
    
  for display_str in display_str_list[::-1]: #for문을 사용해 각 bounding box에 그려질 class명과 그 확률을 font로 나타낸다.
    text_width, text_height = font.getsize(display_str)
    margin = np.ceil(0.05 * text_height)
    draw.rectangle([(left, text_bottom - text_height - 2 * margin),
                    (left + text_width, text_bottom)],
                   fill=color) # 좌표 인자를 사용해서 rectangle 그리기
    draw.text((left + margin, text_bottom - text_height - margin),
              display_str,
              fill="black",
              font=font) # 좌표인자와 display_str 인자를 사용해서 text 삽입
    
    text_bottom -= text_height - 2 * margin


def draw_boxes(image, boxes, class_names, scores, max_boxes=10, min_score=0.1):
    '''
    Draw boxes 함수는 앞서 작성한 Draw_bounding_box_on_image을 사용하여 실제로 image에 bounding box를 그려줄 함수이다.
    Try- except 문을 사용해서 font를 가져오고 boxes라는 인자로 받은 list를 for문으로 꺼낸다.
    boxes라는 인자에는 model의 결과로 받은 output 중 예측된 bounding box의 좌표가 입력된다.
    Confidence score가 0.1보다 높으면 bounding box를 그린다. 
    Display_str에 class명과 그 score를 함께 적어서 bounding box가 어떻게 inference 하고 있는지 쉽게 이해할 수 있도록 한다.
    이제 모든 전처리는 끝나고 pretrain된 SSD detector를 사용해서 output값을 얻어서 위의 함수를 사용하면,
    최종 bounding box가 그려진 image를 얻을 수 있다.
    
    '''
  colors = list(ImageColor.colormap.values()) # color 지정

  try:
    font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationSansNarrow-Regular.ttf",
                              25)
  except IOError:
    print("Font not found, using default font.")
    font = ImageFont.load_default()

  for i in range(min(boxes.shape[0], max_boxes)):
    if scores[i] >= min_score:
      ymin, xmin, ymax, xmax = tuple(boxes[i])
      display_str = "{}: {}%".format(class_names[i].decode("ascii"),
                                     int(100 * scores[i]))
      color = colors[hash(class_names[i]) % len(colors)]
      image_pil = Image.fromarray(np.uint8(image)).convert("RGB")
      draw_bounding_box_on_image(
          image_pil,
          ymin,
          xmin,
          ymax,
          xmax,
          color,
          font,
          display_str_list=[display_str])
      np.copyto(image, np.array(image_pil))
  return image
```

```python
downloaded_image_path = resize_image("convert.JPG", "sample.JPG", 1280, 856, True)
# image 저장 및 출력,
# 해당 이미지는 downloaded_image_path에 저장된다.
```

```python
module_handle = "https://tfhub.dev/google/openimages_v4/ssd/mobilenet_v2/1"

#Tensorflow Hub에 있는 pretrain된 SSD 모델을 불러온다. 
#해당 SSD 모델은 mobilenet_v2를 backbone으로 사용한 detector이다.

detector = hub.load(module_handle).signatures['default']
#Tensorflow hub에서 제공하는 메서드로 SSD를 load한 후 detector라는 변수에 저장한다.

def load_img(path):
    '''
    Load_img라는 함수는 맨 앞에서 전처리된 image의 path값으로 image를 불러오는 함수이다.
    Return은 전처리 된 image이다.
    '''
  img = tf.io.read_file(path)
  img = tf.image.decode_jpeg(img, channels=3)
  return img

def run_detector(detector, path):
    '''
    Run_detector 함수는 pretrained SSD로 input image를 inference하는 함수이다. 
    Result = detector(converted_img) 라는 코드로 detector가 반환하는 output을 result라는 변수에 저장한다. 
    그 다음에 앞서 만든 bounding box를 그려주는 함수인 draw_boxes에 알맞게 인덱싱 한 결과를 넣어준다.
    '''
  img = load_img(path)

  converted_img  = tf.image.convert_image_dtype(img, tf.float32)[tf.newaxis, ...]
  start_time = time.time()
  result = detector(converted_img)
  end_time = time.time()

  result = {key:value.numpy() for key,value in result.items()}

  print("Found %d objects." % len(result["detection_scores"]))
  print("Inference time: ", end_time-start_time)

  image_with_boxes = draw_boxes(
      img.numpy(), result["detection_boxes"],
      result["detection_class_entities"], result["detection_scores"])
  display_image(image_with_boxes)    
    


```

```python
run_detector(detector, downloaded_image_path)
#Run_detector를 실행시킨 결과이다.
```
