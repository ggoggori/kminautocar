# image를 frame으로 전환

```python
from os.path import isfile, join
import os
import cv2
from google.colab.patches import cv2_imshow

def make_frame(video_path):
    os.makedirs('frame folder', exist_ok=True)
    vidcap = cv2.VideoCapture(video_path)
    success,image = vidcap.read()
    count = 0
    while success:
        cv2.imwrite("frame folder/%d.jpg" % count, image)     # save frame as JPEG file      
        success,image = vidcap.read()
        count += 1
make_frame('/content/drive/MyDrive/ColabNotebooks/차량지능기초/과제 4/dsd.mp4')
```

# frame에서 face detect

```python
import mediapipe as mp
mp_face_detection = mp.solutions.face_detection

# Prepare DrawingSpec for drawing the face landmarks later.
mp_drawing = mp.solutions.drawing_utils 
drawing_spec = mp_drawing.DrawingSpec(thickness=1, circle_radius=1)


def convert_image(image):
    with mp_face_detection.FaceDetection(
        min_detection_confidence=0.88) as face_detection:
        # Convert the BGR image to RGB and process it with MediaPipe Face Detection.
        results = face_detection.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        annotated_image = image.copy()
        if results.detections is None:
            return image
        for detection in results.detections:
            mp_drawing.draw_detection(annotated_image, detection)
    return annotated_image
```

```python
os.makedirs('converted frame folder', exist_ok=True)
count = 0
frame_list = os.listdir('frame folder')
frame_list.sort(key = lambda x: int(x[:-4]))
for frame in frame_list:
    image = cv2.imread(os.path.join('frame folder',frame))
    converted_image = convert_image(image)
    cv2.imwrite(os.path.join('converted frame folder',"{}.jpg".format(count)), converted_image)
    print(count, 'frame inference complete')
    count += 1
```

# face detection이 완료된 frame 다시 영상으로 합치기

```python
print('frame to video!!')
pathIn= 'converted frame folder'
pathOut = 'video3.mp4'
fps = 20
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
files.sort(key = lambda x: int(x[:-4]))

for i in range(len(files)):
    filename= os.path.join(pathIn,files[i])
    #reading each files
    img = cv2.imread(filename)
    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'MP4V'), fps, size)
for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])
out.release()
```
